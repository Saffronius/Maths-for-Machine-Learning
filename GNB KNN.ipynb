```python
#Import requisite Libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
   
    
#Initialize Dataframe with raw data
df = pd.read_csv('/Users/adarshvatsa/Downloads/BP.csv')
#Boxplot for age
plt.figure()
sns.boxplot(y=df['Age'])
plt.show()
#Boxplot for height
plt.figure()
sns.boxplot(y=df['Height'])
plt.show()
#Boxplot for weight
plt.figure()
sns.boxplot(y=df['Weight'])
plt.show()
#Pairplots for Age, Height and Weight
sns.pairplot(df[['Age', 'Height', 'Weight']])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score

#Ready Data
X = df.drop('BP_Status', axis=1)
y = df['BP_Status']

#Changing all textual data to numerical
X['Somker'] = X['Somker'].astype('category').cat.codes
X['Gender'] = X['Gender'].map({'Female': 0, 'Male': 1})

#Split data into training(70%) and test(30%) as asked
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#Cart model implementation in python
cart_model = DecisionTreeClassifier()
cart_model.fit(X_train, y_train)

#Make predictions on the test set using CART
y_pred = cart_model.predict(X_test)

#Getting accuracy,F1 and recall 
accuracy_cart = accuracy_score(y_test, y_pred)
f1_cart = f1_score(y_test, y_pred, average='weighted')
recall_cart = recall_score(y_test, y_pred, average='weighted')

print(f'Accuracy: {accuracy_cart}')
print(f'F1 Score: {f1_cart}')
print(f'Recall: {recall_cart}')
```
